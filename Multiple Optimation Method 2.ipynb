## Single Variable Function (Gradient)

Pada tugas ini, berikut Function One Variable yang dipilih:
$$f(x) = 4x^3$$

## Newton - Raphson Method

```python
# Mendefiniskan Fungsi
def f(x):
    return 4*x**3

# Mendefinisikan Fungsi Turunan Pertama
def g(x):
    return 12*x**2

# Implementasi Metode Newton - Raphson

# Mendefinisikan metode newton - raphson dan membuat kalang sesuai algoritma
def newtonRaphson(x0,e,N):
    print('\n\n=== METODE NEWTON - RAPHSON ===')
    step = 1                    # Nilai awal step. Maksimum step diinput oleh pengguna  
    flag = 1                    #
    condition = True            # Kondisi proses berjalan
    while condition:
        if g(x0) == 0.0:
            print('Divide by zero error!')
            break
        
        x1 = x0 - f(x0)/g(x0)   # Kalkulasi metode
        print('Iterasi-%d, x1 = %0.3f dan f(x1) = %0.3f' % (step, x1, f(x1)))
        x0 = x1
        step = step + 1         # Counter up
        
        if step > N:
            flag = 0
            break
        
        condition = abs(f(x1)) > e   # Jika f(x) <= e -> Terminate!
    
    if flag==1:
        print('\nRequired root is: %0.5f' % x1)
    else:
        print('\nNot Convergent.')


# Input oleh pengguna
x0 = float(input('Masukan tebakan: '))
e = float(input('Toleransi Error: '))
N = int(input('Maksimum Step: '))

# Memulai metode newton - raphson setelah mendefinisikan sebelumnya
newtonRaphson(x0,e,N)
```

## Bisection Method

```python
# Mendefinisikan fungsi single variable
def f(x):
    return 4*x**3

# Implementasi Metode Bisection
def bisection(x0,x1,e):           # x0 = a; x1 = b; e = error
    print('\n\n=== IMPLEMENTASI METODE BISECTION ===')
    step = 1                      # step awal iterasi
    condition = True              # Kondisi proses berjalan           
    while condition:
        x2 = (x0 + x1)/2          # x2 = z
        print('Iterasi-%d, x2 = %0.3f dan f(x2) = %0.3f' % (step, x2, f(x2)))

        if f(x0) * f(x2) < 0:
            x1 = x2
        else:
            x0 = x2
        
        step = step + 1               # Counter Up
        condition = abs(f(x2)) > e    # Jika f(z) < e => Terminate!

    print('\nRequired Root is : %0.5f' % x2)


# Input oleh Pengguna
x0 = float(input('Tebakan Pertama: '))
x1 = float(input('Tebakan Kedua: '))
e = float(input('Toleransi Error: '))

# Memeriksa Kebenaran nilai tebakan awal dan membagi dua
if f(x0) * f(x1) > 0.0:
    print('Given guess values do not bracket the root.')
    print('Try Again with different guess values.')
else:
    bisection(x0,x1,e)
```

## Secant Method

```python
# Mendefinisikan fungsi awal
def f(x):
    return 4*x**2

# Implementasi Metode Secant

def secant(x0,x1,e,N):
    print('\n\n=== IMPLEMENTASI METODE SECANT ===')
    step = 1                                    # step awal iterasi
    condition = True                            # Kondisi proses berjalan 
    while condition:
        if f(x0) == f(x1):
            print('Divide by zero error!')
            break
        
        x2 = x0 - (x1-x0)*f(x0)/( f(x1) - f(x0) ) # x0 = a; x1 = b; x2 = z
        print('Iterasi-%d, x2 = %0.3f dan f(x2) = %0.3f' % (step, x2, f(x2)))
        x0 = x1
        x1 = x2
        step = step + 1                         # Counter Up
        
        if step > N:                           
            print('Not Convergent!')
            break
        
        condition = abs(f(x2)) > e              # Jika f(z) < e => Terminate!
    print('\nRequired root is: %0.5f' % x2)


# Input oleh Pengguna
x0 = float(input('Masukan Tebakan Pertama: '))
x1 = float(input('Masukan Tebakan Kedua: '))
e = float(input('Toleransi Error: '))
N = int(input('Maksimum Step: '))

# Starting Secant Method
secant(x0,x1,e,N)
```
